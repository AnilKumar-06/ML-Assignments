```python
import numpy as np

def gradient_descent(x, y):
    m_curr = b_curr = 0
    itr = 1000
    lr = 0.001
    n = len(x)
    for i in range(itr):
        y_pred = m_curr*x + b_curr
        cost = (i/n)*sum([val**2 for val in (y-y_pred)])
        md = -(2/n)*(sum(x*(y-y_pred)))
        bd = -(2/n)*(sum(x*(y-y_pred)))
        m_curr = m_curr - lr * md
        b_curr = b_curr - lr * bd
        print("m===>{},    b===>{},    cost===>{},    itr===>{}".format(m_curr, b_curr, cost, i))

x = np.array([1,2,3,4,5,])
y = np.array([5,7,9,11,13])

gradient_descent(x, y)
```
